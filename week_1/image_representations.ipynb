```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from os.path import join
```

First we open a grayscale image, which will have dimension (Height x Width) as there is only a single color value for each pixel representing the intensity.

```python
image_path_gray = join('images', 'image_gray.jpg')
image_gray = cv2.imread(image_path_gray, cv2.IMREAD_GRAYSCALE)
plt.imshow(image_gray, cmap='gray')
plt.axis('off')
plt.show()
print(image_gray.shape)
```

```python
image_path = join('images', 'img_1.jpg')
image_1 = cv2.imread(image_path)
image_1_rgb = cv2.cvtColor(image_1, cv2.COLOR_BGR2RGB)
print(type(image_1), type(image_1_rgb))
```

```python
fig, axs = plt.subplots(nrows=1, ncols=2)

axs[0].imshow(image_1)
axs[0].set_title('BGR Image')
axs[0].axis('off')
axs[1].imshow(image_1_rgb)
axs[1].set_title('RGB Image')
axs[1].axis('off')
plt.show()
```

Now we make a function that shows the distribution of pixels in each color channel

```python
def show_color_distribution(image):
    """
    Display the distribution of pixels in each color channel (RGB) of an image.
    
    Parameters:
    - image: NumPy array representing the image.
    """
    # Split the image into its color channels
    colors = ('b', 'g', 'r')
    channel_names = ('Blue', 'Green', 'Red')
    
    fig, axs = plt.subplots(nrows=1, ncols=4, figsize=(18, 6))
    axs[0].set_title('Image')
    axs[0].imshow(image)
    axs[0].axis('off')
    
    for i, (color, channel_name) in enumerate(zip(colors, channel_names)):
        axs[i + 1].hist(image[..., i].flatten(), bins=256, color=color, alpha=0.7)
        axs[i + 1].set_title(f'{channel_name} Channel')
        axs[i + 1].set_xlabel('Pixel Intensity')
        axs[i + 1].set_ylabel('Frequency')
        axs[i + 1].set_xlim([0, 256])
    
    plt.tight_layout()
    plt.show()
```

```python
show_color_distribution(image_1_rgb)
```
